```python
import gym
import safe_agents as sa
```

```python
EPISODES = 100
env = gym.make('LunarSafe-v0')
```

```python
baseline = sa.agents.BaselineAgent(env)
scores1, safety1 = baseline.train(episodes=EPISODES)
```

```python
dqn = sa.agents.DQNAgent(env)
scores2, safety2 = dqn.train(episodes=EPISODES)
```

```python
dqn_control = sa.agents.DQNControlAgent(env)
scores3, safety3 = dqn_control.train(episodes=EPISODES)
```

```python
a2c = sa.agents.A2CAgent(env)
scores4, safety4 = a2c.train(episodes=EPISODES)
```

```python
a2c_control = sa.agents.A2CControlAgent(env)
scores5, safety5 = a2c_control.train(episodes=EPISODES)
```

```python
agents = ["baseline", "dqn", "dqn_control", "a2c", "a2c_control"]
scores = scores1 + scores2 + scores3 + scores4 + scores5
sa.utils.plot_multi_scores(agents, scores)
```

```python
data = {'baseline': safety1, 'dqn': safety2, 'dqn_control': safety3, 'a2c': safety4, 'a2c_control': safety5}
sa.utils.plot_multi_safety(data)
```
